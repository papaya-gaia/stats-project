```python
import pandas as pd
housing = pd.read_csv('housing.csv')
print(housing.head())
```

```python
print(housing.info())
```

```python
pd.DataFrame(housing)
```

```python
#Descriptive statistics for all columns
print(housing.describe(include='all'))
```

```python
#Missing values
missing_per_column = housing.isnull().sum()
missing_per_column
```

```python
#See if there are missing values in Price column
missing_price = housing['Price'].isna().sum()
print(f'Number of missing values in the Price column: {missing_price}')

price_is_zero=(housing['Price']==0).sum()
print(f'Number of rows with Price equal to 0: {price_is_zero}')
```

```python
# Average price per Seller
pd.set_option('display.float_format','{:,.2f}'.format) #avoid scientific notation
mean_price_per_seller = (housing.groupby('SellerG')['Price']
    .mean()
    .sort_values(ascending=False)
    .fillna(0)
    .reset_index())
pd.DataFrame(mean_price_per_seller)
```

```python
#How many different sellers are there?
#Should be 268
seller_count = housing['SellerG'].value_counts().reset_index()
pd.DataFrame(seller_count)
```

```python
#How many different regions are there?
region_count = housing['Regionname'].value_counts().reset_index()
pd.DataFrame(region_count)
```

```python
#Average price per Region
mean_price_per_region = (housing.groupby('Regionname')['Price']
                         .mean()
                         .sort_values(ascending=False)
                         .reset_index()
                         )
pd.DataFrame(mean_price_per_region)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Extract target variable
price = housing['Price']

#Check skewness
skewness_before = price.skew()
print(f"Skewness Before Transformation: {skewness_before}")
```

```python
#Apply log transformation using numpy
log_price = np.log(price)
skewness_after = log_price.skew()
print(f"Skewness after transformation: {skewness_after}")
```

```python
#Plotting the histograms for the price before and after log transformation
plt.figure(figsize=(12,6)) #set the figure size

#Original Price distribution
plt.subplot(1,2,1)
plt.hist(price,bins=30,alpha=0.7,color='blue')
plt.title('Original Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.axvline(price.mean(),color='red',linestyle='dashed',linewidth=1,label='Mean')
plt.legend()

#Log transformed Price distribution
plt.subplot(1,2,2)
plt.hist(log_price,bins=30,alpha=0.7,color='green')
plt.title('Log-transformed Price Distribution')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')
plt.axvline(log_price.mean(),color='red',linestyle='dashed',linewidth=1,label='Mean')
plt.legend()

plt.tight_layout()
plt.show()
```
